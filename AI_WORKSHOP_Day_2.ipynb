```python
from PIL import Image,ImageEnhance
import numpy as np
def preprocess_image(image_path,size=(128,128),rotate_angle=0,brightness=1.0,contrast=1.0):
  image=Image.open(image_path)
  image=image.resize(size)
  image=image.rotate(rotate_angle)
  image=ImageEnhance.Contrast(image).enhance(contrast)
  image=ImageEnhance.Brightness(image).enhance(brightness)
  image=np.array(image)
  image=image/255.0 #scaling
  return image



```

```python
image_folder="/content/pokemon/images"
processed_image=[]
for file in os.listdir(image_folder):
  image_path=os.path.join(image_folder,file)
  img=preprocess_image(image_path,size=(128,128),rotate_angle=10,brightness=1.2,contrast=1.5)
  processed_image.append(img)

processed_image=np.array(processed_image)
print(processed_image.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(processed_image[i])
    plt.axis("off")

plt.show()
```
